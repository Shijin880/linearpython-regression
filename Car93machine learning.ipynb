```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df_car.isnull().sum()
```

```python
df_car['AirBags'].fillna(df_car['AirBags'].mode(0),inplace=True)
df_car['Luggage.room'].fillna(df_car['Luggage.room'].mode(0),inplace=True)
```

```python
df_car.isnull().sum()
```

```python
df_car.info()
```

```python
df_car['Cylinders'].unique()
```

```python
df_car['Cylinders'].replace('rotary',df_car['Cylinders'].mode()[0],inplace=True)
```

```python
df_car['Cylinders'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X= df_car[['Horsepower']]
Y=df_car['Price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_predict= model.predict(X_test)
```

```python
mse= mean_squared_error(Y_test,Y_predict)
```

```python
mse
```

```python
from sklearn.metrics import r2_score
r2 =r2_score(Y_test,Y_predict)
```

```python
r2
```

```python
plt.scatter(X_test,Y_test,color='blue',label='Actual Price')
plt.scatter(X_test,Y_predict,color='red',label='Predicted Price')
plt.plot(X_test,Y_predict,color='Green',label='Regression Line')
plt.xlabel('features')
plt.ylabel('Price')
plt.legend()
plt.suptitle(f'Regression of line with r2:{r2:.2f}')
plt.show()
```
