```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X=df[['pizza_size']]
Y=df[['unit_price']]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
X['pizza_size']=le.fit_transform(X['pizza_size'])
```

```python
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
X_train_scaled=scale.fit_transform(X_train)
X_test_scaled=scale.transform(X_test)
```

```python
model.fit(X_train_scaled,Y_train)
```

```python
Y_pred= model.predict(X_test_scaled)
```

```python
mse= mean_squared_error(Y_test,Y_pred)
r2=r2_score(Y_test,Y_pred)
mse
r2
```

```python
mse
```

```python
plt.scatter(X_test,Y_test,color='blue',label="Actual Price")
plt.scatter(X_test,Y_pred,color='Red',label='Predicted Price')
plt.plot(X_test,Y_pred,color='Green',label='Regression Line')
plt.xlabel('Pizza Size')
plt.ylabel('Price')
plt.title('Pizza Size vs Price')
plt.suptitle(f'Regression of Line with r2:{r2:.2f}')
plt.legend()
plt.show()
```
